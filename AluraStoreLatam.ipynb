### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos=lambda x: x['Precio'].sum()
```

```python
#Ingresto total de la tienda 1
ingresos_1=ingresos(tienda)

#ingresos totales de la tienda 2
ingresos_2=ingresos(tienda2)

#ingresos totales de la tienda 3
ingresos_3=ingresos(tienda3)

#ingresos totales de la tienda 4
ingresos_4=ingresos(tienda4)

#print
print(f'Los ingresos de las tiendas de Jao son los siguientes: tienda 1:{ingresos_1}, tienda 2:{ingresos_2}, tienda 3:{ingresos_3} y tienda 4:{ingresos_4}')
```

```python
#Tabla resumen de lo ingresos de tiendas
tiendas_ingresos={'tienda 1':1150880400.0,
                  'tienda 2':1116343500.0,
                  'tienda 3':1098019600.0,
                  'tienda 4':1038375700.0}
```

```python
#para saber la tienda con menos ventas
min_ingresos=min(tiendas_ingresos.values())
nombre_tienda=min(tiendas_ingresos, key=tiendas_ingresos.get)
print(f'La tienda con menos ingresos es la tienda {nombre_tienda} con {min_ingresos} de ingreso')
```

```python
#ingresos totales de las 4 tiendas alura
total=ingresos_1+ingresos_2+ingresos_3+ingresos_4

#
porcentaje_ingresos= lambda x: round((x/total)*100,2)

#tienda 1
porcentaje_ingresos_1=porcentaje_ingresos(ingresos_1)

#tienda 2
porcentaje_ingresos_2=porcentaje_ingresos(ingresos_2)

#tienda 3
porcentaje_ingresos_3=porcentaje_ingresos(ingresos_3)

#tienda 4
porcentaje_ingresos_4=porcentaje_ingresos(ingresos_4)
```

```python
#gráfica de pastel que representa cuantos ingresos del total genero cada tienda
import matplotlib.pyplot as plt

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [porcentaje_ingresos_1, porcentaje_ingresos_2, porcentaje_ingresos_3, porcentaje_ingresos_4]

# Crear el gráfico de pastel
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%')
plt.title('Porcentaje de ingresos por tienda')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio=lambda x: round(sum(x['Calificación'])/len(x['Calificación']),2)
```

```python
#tienda 1 calificación promedio
calificacion1=promedio(tienda)

#tienda 2 calificación promedio
calificacion2=promedio(tienda2)

#tienda 3 calificación promedio
calificacion3=promedio(tienda3)

#tienda 4 calificación promedio
calificacion4=promedio(tienda4)

#print
print(f'La calificación promedio de la tienda 1 es {calificacion1}, de la tienda 2 es {calificacion2}, de la tienda 3 es {calificacion3} y de la tienda 4 es {calificacion4}')
```

```python
#gráfica de barras para las calificaciones de las 4 tiendas
import matplotlib.pyplot as plt

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion1, calificacion2, calificacion3, calificacion4]

# Crear el gráfico de barras
plt.bar(tiendas, calificaciones)
plt.xlabel('Tiendas')
plt.ylabel('Calificación promedio')
plt.title('Calificación promedio por tienda')
plt.show()
```

# 5. Envío promedio por tienda

```python
#función lambda para el promedio por tienda
promedio_envio=lambda x: round(sum(x['Costo de envío'])/len(x['Costo de envío']),2)
```

```python
#costo de envio tienda 1
costo_envio_t1=promedio_envio(tienda)

#costo de envio tienda 2
costo_envio_t2=promedio_envio(tienda2)

#costo de envio tienda 3
costo_envio_t3=promedio_envio(tienda3)

#costo de envio tienda 4
costo_envio_t4=promedio_envio(tienda4)

#print
print(f'Los costos de envio para la tienda 1 son {costo_envio_t1}, para la tienda 2 son de {costo_envio_t2}, para la tienda 3 son de {costo_envio_t3} y para la tienda 4 son de {costo_envio_t4}')
```

```python
#boxplot de los costos de envio de las 4 tiendas
import matplotlib.pyplot as plt

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [tienda['Costo de envío'], tienda2['Costo de envío'], tienda3['Costo de envío'], tienda4['Costo de envío']]

# Crear el gráfico de caja
plt.boxplot(costos_envio)
plt.xticks([1, 2, 3, 4], tiendas)
plt.xlabel('Tiendas')
plt.ylabel('Costo de envío')
plt.title('Costo de envío por tienda')
plt.show()
```
